### PREPROCESSING

```python
import pandas as pd
```

```python
dataset = pd.read_csv('poli_texts.csv')
```

```python
len(dataset)
```

```python
dataset.head(10)
```

```python
import pandas as pd
import re
import nltk
from sklearn.model_selection import train_test_split
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.preprocessing import LabelEncoder
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
if 'Unnamed: 0' in dataset.columns:
    dataset.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
dataset.head(5)
```

```python
# missing values

dataset.dropna(subset=['text'], inplace=True)
dataset.dropna(subset=['username'], inplace=True)
dataset.dropna(subset=['party'], inplace=True)
```

```python
len(dataset)
```

```python
# text to lowercase

dataset['text'] = dataset['text'].str.lower()
dataset['username'] = dataset['username'].str.lower()
dataset['party'] = dataset['party'].str.lower()
```

```python
dataset.head(5)
```

```python
# remove urls
dataset['text'] = dataset['text'].apply(lambda x: re.sub(r'http\S+', '', x))
```

```python
dataset.head(5)
```

```python
# remove punctuation and special chars
dataset['text'] = dataset['text'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
```

```python
dataset.head(5)
```

```python
# remove stopwords
stop_words = set(stopwords.words('english'))
dataset['text'] = dataset['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
```

```python
dataset.head(5)
```

```python
dataset.to_csv('poli_texts_processed.csv')
```
